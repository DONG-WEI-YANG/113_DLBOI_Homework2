Numpy is the core library for scientific computing in Python. It provides a high-performance multidimensional array object and tools for working with these **arrays**.

### **Arrays**

- A numpy array is a grid of values, all of the same type, and is indexed by a tuple of nonnegative integers.

- The number of dimensions is the rank of the array;

- The shape of an array is a tuple of integers giving the size of the array along each dimension.

### **Array Operations**

```python
# create 2 2d-array
import numpy as np
x = np.array([[1, 2],[3, 4]], dtype=int)
y = np.array([[5, 6],[7, 8]], dtype=int)
# Elementwise sum; both produce an array
print(x + y)
print(np.add(x, y))
```

```python
# Elementwise difference; both produce an array
import numpy as np
x = np.array([[1, 2],[3, 4]], dtype=int)
y = np.array([[5, 6],[7, 8]], dtype=int)
print(x - y)
print(np.subtract(x, y))
```

```python
# Elementwise division; both produce an array
import numpy as np
x = np.array([[1, 2],[3, 4]], dtype=int)
y = np.array([[5, 6],[7, 8]], dtype=int)
print(x / y)
print(np.divide(x, y))
```

```python
# Elementwise product; both produce the array
import numpy as np
x = np.array([[1, 2],[3, 4]], dtype=int)
y = np.array([[5, 6],[7, 8]], dtype=int)
print(x * y)
print(np.multiply(x, y))
```

```python
# Dot product or matrix multiplication
import numpy as np
x = np.array([[1, 2],[3, 4]], dtype=int)
y = np.array([[5, 6],[7, 8]], dtype=int)
print(x.dot(y))
print(np.dot(x, y))
```

```python
# Denoise by filter
import cv2
import matplotlib.pyplot as plt
import numpy as np

filename = 'hw2.png'  # your file path
img = cv2.imread(filename)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for correct color display
(h, w, d) = img.shape  # Read picture format
center = (w // 2, h // 2)  # Find the center of image
P = cv2.getRotationMatrix2D(center, -180, 1.0)
rotate_img = cv2.warpAffine(img_rgb, P, (w, h))
img_raw=rotate_img
denoised_image = cv2.GaussianBlur(img_raw, (21, 21), 0)
output_filename = 'denoised_image.png'
output_filename = 'denoised_image.png'
cv2.imwrite(output_filename, cv2.cvtColor(denoised_image, cv2.COLOR_RGB2BGR))
plt.imshow(denoised_image)
plt.title('Denoised Image')
plt.show()
```
